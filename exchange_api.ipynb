```python
import requests
import pandas as pd
import datetime
```

```python
api_key = 'YOUR_API_KEY' #For getting api visit oenexchange.org
url = 'https://openexchangerates.org/api/historical/2013-02-16.json'
```

```python
target = url+'?'+'app_id='+api_key
req = requests.get(target)
```

```python
resp = req.json()
```

```python
resp
```

```python
resp['rates']
```

```python
df = pd.DataFrame(resp['rates'].items(), columns=['currency_name', 'currency_rate'])
df['curreny_date'] = datetime.datetime.fromtimestamp(resp['timestamp'])
df['imported_date'] = pd.to_datetime('now')
```
