```python
import requests
```

```python
api_key = 'YOUR_API_KEY' #For getting API key visit api.themoviedb.org
url = 'https://api.themoviedb.org/3/movie/popular?api_key='+api_key
```

```python
req = requests.get(url).json()
```

```python
req
```

```python
for result in req['results']:
    print(f"Movie Id: {result['id']}")
    print(f"Original Title : {result['original_title']}")
    print(f"Movie Title: {result['title']}")
    print(f"For Adults ? : {result['adult']}") 
    print(f"Genre Ids : {result['genre_ids']}")
    print(f"Original Language : {result['original_language']}")
    print(f"Overview : {result['overview']}")
    print(f"Popularity : {result['popularity']}")
    print(f"Release Date : {result['release_date']}")
    print(f"Vote Average : {result['vote_average']}")
    print(f"Vote Count : {result['vote_count']}")
```

```python
url_tv = 'https://api.themoviedb.org/3/tv/popular?api_key='+api_key
req_tv = requests.get(url_tv).json()
```
